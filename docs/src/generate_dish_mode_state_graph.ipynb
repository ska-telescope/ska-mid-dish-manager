```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from ska_mid_dish_manager.models.dish_mode_model import DishModeModel

dm_graph = DishModeModel().dishmode_graph

pos = nx.spring_layout(dm_graph, seed=3, k=1.2)

# Create distinct edge colors
edges = list(dm_graph.edges())
n_edges = len(edges)
colors = plt.cm.viridis(np.linspace(0, 1, n_edges))

# Draw edges
fig, ax = plt.subplots(figsize=(10, 7))
nx.draw_networkx_edges(
    dm_graph,
    pos,
    width=2,
    alpha=0.3,
    arrows=True,
    arrowstyle="-|>",
    arrowsize=18,
    edge_color=colors,
)

# Draw nodes
nx.draw_networkx_nodes(
    dm_graph,
    pos,
    edgecolors="black",
    linewidths=1.5,
)

# Draw labels
nx.draw_networkx_labels(
    dm_graph,
    pos,
    font_size=10,
    # font_color="white",
    font_weight="bold",
)

# Plot
ax.set_title("Dish Mode State Graph", fontsize=14, fontweight="bold", pad=20)
ax.axis("off")
fig.tight_layout()
```
